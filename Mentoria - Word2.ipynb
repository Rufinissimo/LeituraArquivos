### Criar um documento

```python
from docx import Document

    # Criar o documento:
documento = Document()

    # Criar o texto a ser inserido:
texto = """Dolly! Dolly Guaraná! Dolly!
(O melhor)
Dolly Guaraná, o sabor Brasileiro!
"""
    # Inserir o texto no documento:
paragrafo = documento.add_paragraph(texto)
```

### Criar diferentes formatações de texto

```python
   # Consultar todos os estilos já existentes no Word:
for estilo in documento.styles:
    print(estilo)
```

```python
   # Importar valores de formatação:
from docx.shared import Pt, RGBColor, Cm # Pt -> permite determinar o tamanho da fonte; RGBColor > permite determinar a cor da fonte; Cm ->permite determinar os valores em centímetro, como margem, por exemplo.
from docx.enum.style import WD_STYLE_TYPE # Permite criar um estilo de texto no word.

paragrafo.style = documento.styles.add_style("Estilo Dolly", WD_STYLE_TYPE.PARAGRAPH) # PARAGRAPH -> argumento para formatação; Consultar os demais argumentos.
paragrafo.style.font.name = "Comic Sans" # Determinar o estilo da fonte.
paragrafo.style.font.size = Pt(13) # Determinar o tamanho da fonte.
paragrafo.style.font.bold = True # Determinar a fonte em negrito.
paragrafo.style.font.italic = True # Determinar a fronte em itálico.
paragrafo.style.font.underline = True # Determinar a fonta sublinhada.
paragrafo.style.font.color.rgb = RGBColor(255 , 0, 0) # Determinada a cor da fonte.

""" Consultar as demais formatações de texto """
```

### Adicionar um texto com edição dentro do texto

```python
documento = Document()

   # Criar uma variável que receba uma determinada informação:
faturamento = 15.476

texto = """Rodrigo, você está contratado.

Seu salário será """

texto_final = "Seja bem-vindo!"

paragrafo = documento.add_paragraph(texto)
paragrafo.add_run(f"R${faturamento}.").bold = True # É possível determinar uma formatação ou um estilo já criado (opcional)
paragrafo = documento.add_paragraph(texto_final)

documento.save("Estou Empregado.docx")
```

### Controle de margens e seções do documento

```python
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

documento = Document()

   # Determinar os valores para margem:
for secao in documento.sections:
    secao.top_margin = Cm(2)
    secao.bottom_margin = Cm(1)
    secao.left_margin = Cm(1.5)
    secao.right_margin = Cm(1.5)

texto = """Mais texto.
Dessa vez, para aplicar o controle de margens e seções.
Pois é! Pois é! Pois é!
"""

paragrafo = documento.add_paragraph(texto)

   # Alinhar o texto seguindo um padrão:
paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER # Consultar outros padrões de alinhamento.

documento.save("Estou Empregado.docx")
```

### Inserir imagem no documento

```python
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

documento = Document()

   # Determinar os valores para margem:
for secao in documento.sections:
    secao.top_margin = Cm(2)
    secao.bottom_margin = Cm(1)
    secao.left_margin = Cm(1.5)
    secao.right_margin = Cm(1.5)

texto = """Mais texto.
Dessa vez, para aplicar o controle de margens e seções.
Pois é! Pois é! Pois é!
"""

paragrafo = documento.add_paragraph(texto)
   
   # Adicionar uma imagem ao documento; determinar os valores de tamamnho da imagem:
documento.add_picture("imagem.png", width=Cm(4), height=Cm(4)) # Se o arquivo de imagem não estiver na mesma pasta do documento, determinar o caminho da imagem.

documento.save("Estou Empregado.docx")
```

### Inserir tabela no documento

```python
documento = Document()

   # Criar uma base de dados:
records = (
    (3, '101', 'Alho'),
    (7, '422', 'Cebola'),
    (4, '631', 'Alho, Alho, Cebola, e Alho')
)

   # Criar uma tabela:
table = documento.add_table(rows=1, cols=3, style="Light Grid Accent 1")
hdr_cells = table.rows[0].cells # Determinando o cabeçalho da tabela.
hdr_cells[0].text = 'Quantidade'
hdr_cells[1].text = 'Id'
hdr_cells[2].text = 'Descrição'
for qty, id, desc in records:
    row_cells = table.add_row().cells
    row_cells[0].text = str(qty)
    row_cells[1].text = id
    row_cells[2].text = desc
    
    
documento.save("Texto.docx")
```

### Adaptar um documento

```python
from datetime import datetime

contrato = Document("Contrato.docx")

nome = "Lira da Hashtag"
item1 = "Serviço de Treinamento em Excel"
item2 = "Apostila Completa de Excel"
item3 = "Serviço de Treinamentos de Python"

dicionario_valores = {
    "XXXX": nome,
    "YYYY": item1,
    "ZZZZ": item2,
    "WWWW": item3,
    "DD": str(datetime.now().day), # Retornar o dia atual.
    "MM": str(datetime.now().month), # Retornar o mês atual.
    "AAAA": str(datetime.now().year), # Retornar o ano atual.
}

# tem alguma função replace? -> é executada por parágrafo
# percorrer as linhas
    # se a linha tiver o texto xxxx substitui pelo nome
for paragrafo in contrato.paragraphs:
    # para cada placeholder do dicionario
    for codigo in dicionario_valores:
        if codigo in paragrafo.text:
            paragrafo.text = paragrafo.text.replace(codigo, dicionario_valores[codigo])

contrato.save(f"Contrato Atualizado - {nome}.docx")
```
